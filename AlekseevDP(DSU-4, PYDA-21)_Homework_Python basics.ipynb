```python
"""Алексеев Д.П. (DSU-4, PYDA-21)
Домашнее задание к лекции "Основы Python"

Задание 1
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет, какая из этих строк длиннее."""

phrase_1 = input("Введите строку №1:")
phrase_2 = input("Введите строку №2:")

if len(phrase_1) > len(phrase_2):
    print ("Строка №1 больше строки №2")
elif len(phrase_1) < len(phrase_2):
    print ("Строка №1 меньше строки №2")
else:
    print ("Строка №1 равна строке №2")
```

```python
"""Задание 2
Дана переменная, в которой хранится четырехзначное число (год). 
Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Википедия: 
https://ru.wikipedia.org/wiki/%D0%92%D0%B8%D1%81%D0%BE%D0%BA%D0%BE%D1%81%D0%BD%D1%8B%D0%B9_%D0%B3%D0%BE%D0%B4
-год, номер которого кратен 400, — високосный;
-остальные годы, номер которых кратен 100, — невисокосные (например, годы 1700, 1800, 1900, 2100, 2200, 2300);
-остальные годы, номер которых кратен 4, — високосные.
Последние годы столетий, оканчивающиеся на два нуля, в трёх случаях из четырёх не являются високосными. 
Так, годы 1700, 1800 и 1900 не являются високосными, так как они кратны 100 и не кратны 400. 
Годы 1600 и 2000 — високосные, так как они кратны 400. 
Годы 2100, 2200 и 2300 — невисокосные."""

year_input = int(input('Введите год в 4-хзначном формате: '))
if year_input % 4 != 0:
    print(year_input, 'год - невисокосный (обычный)')
elif year_input % 100 != 0:
    print(year_input, 'год - високосный')
elif year_input % 400 != 0:
    print(year_input, 'год - невисокосный (обычный)')
else:
    print(year_input, 'год - високосный')
```

```python
"""Задание 3
Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения 
и выводить соответствующий знак зодиака.

Овен: 21 марта — 20 апреля
Телец: 21 апреля — 20 мая
Близнецы: 21 мая — 21 июня
Рак: 22 июня — 22 июля
Лев: 23 июля — 22 августа
Дева: 23 августа — 22 сентября
Весы: 23 сентября — 22 октября
Скорпион: 23 октября — 21 ноября
Стрелец: 22 ноября — 20 декабря
Козерог: 21 декабря — 19 января
Водолей: 20 января — 19 февраля
Рыбы: 20 февраля — 20 марта
"""

day = int(input("Введите день (числом от 1 до 31):"))
month = int(input("Введите месяц (числом от 1 до 12):"))

if (31>=day>=21 and month==3) or (20>=day>=1 and month==4):
    print('Овен')
elif (30>=day>=21 and month==4) or (20>=day>=1 and month==5):
    print('Телец')
elif (31>=day>=21 and month==5) or (21>=day>=1 and month==6):
    print('Близнецы')
elif (30>=day>=22 and month==6) or (22>=day>=1 and month==7):
    print('Рак')
elif (31>=day>=23 and month==7) or (22>=day>=1 and month==8):
    print('Лев')
elif (31>=day>=23 and month==8) or (22>=day>=1 and month==9):
    print('Дева')
elif (30>=day>=23 and month==9) or (22>=day>=1 and month==10):
    print('Весы')
elif (31>=day>=23 and month==10) or (21>=day>=1 and month==11):
    print('Скорпион')
elif (30>=day>=22 and month==11) or (20>=day>=1 and month==12):
    print('Стрелец')
elif (31>=day>=21 and month==12) or (19>=day>=1 and month==1):
    print('Козерог')
elif (31>=day>=20 and month==1) or (19>=day>=1 and month==2):
    print('Водолей')
elif (29>=day>=20 and month==2) or (20>=day>=1 and month==3):
    print('Рыбы')
else:
    print('Ошибка ввода даты')
```
